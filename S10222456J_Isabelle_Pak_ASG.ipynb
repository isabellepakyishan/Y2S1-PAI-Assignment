Those with loan terms at least 120 months, but less than 300 months will be rated as "Medium". Applicants asking for loans with loan terms of 300 or more months, will be rated as "Long".

Write a function that takes an applicant's numerical value of months of Loan_Term as input parameter and returns the respective customer's rating. Create a new attribute "Loan_Tenure" for every applicant in df_loans.

Display df_loans with only the "Loan_Term" and "Loan_Tenure" attributes.

```python
# df_loan_termAndTenure = df_loans.loc[:, ["Loan_Term"]]

# #write as function
# df_loan_termAndTenure["Loan_Tenure"] = "Short"
# df_loan_termAndTenure.loc[df_loan_termAndTenure["Loan_Term"] < 300, "Loan_Tenure"] = "Medium"
# df_loan_termAndTenure.loc[df_loan_termAndTenure["Loan_Term"] >= 300, "Loan_Tenure"] = "Long"

# df_loan_termAndTenure
df_loans["Loan_Tenure"] = " "

def check_loan_term(loan_term, index):
    if loan_term < 120:
        df_loans.at[index, "Loan_Tenure"] = "Short"
    elif loan_term < 300:
        df_loans.at[index, "Loan_Tenure"] = "Medium"
    else:
        df_loans.at[index, "Loan_Tenure"] = "Long"

for index, row in df_loans.iterrows():
    current_index = index
    loan_term = row.Loan_Term
    check_loan_term(loan_term, current_index)

df_loans.loc[:, ["Loan_Term", "Loan_Tenure"]]
```

### Q2e)  5 marks
Identify the top 2 applicants with generated the Total Income to Loan Ratio, being Highest Combined Income, (both ApplicantIncome and CoapplicantIncome) divided by Loan Amount.

State one possible issue with the dataset.

```python
#create column for total income to loan ratio (applicant and co-applicant income divided by loan amount)
#find top 2 from the ratio
df_loans["Income_Loan_Ratio"] = 0.0

for index, row in df_loans.iterrows():
    combined_income = row.ApplicantIncome + row.CoapplicantIncome
    loan_amount = row.LoanAmount
    income_to_loan = combined_income / loan_amount
    df_loans.at[index, "Income_Loan_Ratio"] = income_to_loan

df_loans.nlargest(n=2, columns=['Income_Loan_Ratio'])
```

One possible issue with the dataset is that the applicant income is much lower than that of the co-applicant. Additionally, the co-applicant's income is more than sufficient to cover the loan amount, which is evident in 
the income-to-loan ratio which is greater than 1, meaning that the total income from the applicant and co-applicant is sufficient to cover the applied loan amount

### Q2f) 10 marks
The column "LoanAmount" details the amount of loan requested for, in Singapore Dollars, S$. 

Targeted marketing will be conducted for the following groups of customers:

1. Loan Amount: >= S\\$50,000 and  < S\\$250,000
2. Loan Amount: >= S\\$250,000 and  < S\\$450,000

Perform basic statistical analysis and comment on the key differences between these two groups of customers.

```python
#low_loan stores the resultant DataFrame from the provided query expression, which finds the loan applications for loan amounts ranging from $50 000 to less than $250 000
low_loan = df_loans.query("LoanAmount >= 50000 and LoanAmount < 250000")

#The .describe() method generates descriptive statistics and calculates statistical data from the DataFrame stored in low_loan
low_loan.describe()
```

```python
#high_loan stores the resultant DataFrame from the provided query expression, which finds the loan applications for loan amounts ranging from $250 000 to less than $450 000
high_loan = df_loans.query("LoanAmount >= 250000 and LoanAmount < 450000")

#The .describe() method generates descriptive statistics and calculates statistical data from the DataFrame stored in low_loan
high_loan.describe()
```

From the above tables displaying the statistical analysis of DataFrames low_loan and high_loan, we can deduce that there is a larger number of people taking loans which are less than $250 000. Many of those taking smaller loans have lower mean applicant ($4862.76 | $12,190.76) and co-applicant incomes ($1463.70 | $2485.79), and similarly have a lower mean loan amount ($130,332.54 | $299,575.76) as compared to those who take larger loans. Those applying for larger loans tend to have medium to long loan terms, while those who applied for smaller loans tend to have loan terms from as low as 36 months, compared to those who applied for larger loans, having loan terms ranging from 180 months to a maximum of 360 months. However, the overall mean income-to-loan ratio for applicants applying for either smaller or larger loans are almost the same (0.0499 | 0.0480)
